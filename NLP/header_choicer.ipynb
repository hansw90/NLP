```
f = open("./similary_matrix_result_0.9_kobert_nvidia.txt", 'r', encoding = 'UTF-8')
```

```
lines = []
# 지금 추가된 글은 다음에 추가되지 않게 하기 위함,
hashLine = []
real_idx = 0
while True :
    line = f.readline()
    split_line = []
    if not line : 
        break
    check_line = line.split(' ')
    split_line = line.split(',')
    if line.startswith('문장 : ') : 
        header_line = line.split('문장 : ')
        if (header_line[-1][:-1] not in hashLine) :
          lines.append([line])
          hashLine.append(header_line[-1][:-1])
        
        continue
    if (len(split_line) > 1) : 
        try :
            if (float(split_line[-1]) > 0.91 and float(split_line[-1]) < 0.98) :

                comma_join = ','.join(split_line[:-1])
                blank_join = ''.join(split_line[:-1])
                if (comma_join not in hashLine and blank_join not in hashLine) :
                    if len(split_line[:-1]) > 1 :
                      lines[len(lines)-1].append(comma_join)
                      hashLine.append(comma_join)
                    else :
                      lines[len(lines)-1].append(blank_join)
                      hashLine.append(blank_join)
                    
        except :
            continue
            #print(line)
    
```
